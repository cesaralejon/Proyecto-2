**Introducción**

¿Qué es el inventario activo? Las empresas tienen una gran variedad de productos en sus existencias, lo que se denomina inventario activo. Aquí tenemos un conjunto de datos de inventario histórico e inventario activo de varios productos. La identidad de los productos se puede extraer del conjunto de datos dado a partir de la columna SKU_number.

Tenemos que predecir el inventario contra las existencias activas mediante la creación de un algoritmo de aprendizaje automático. Estamos dividiendo los inventarios históricos en base al análisis ABC para priorizar los artículos del inventario.


**¿Qué es el análisis ABC?** 


El análisis ABC es un enfoque para clasificar los artículos del inventario en función de los valores de consumo de los artículos. 

El valor de consumo es el valor total de un artículo consumido durante un periodo de tiempo determinado, por ejemplo un año. Vamos a analizar el conjunto de datos dado utilizando el análisis ABC para asesorar sobre las categorías de productos que tendrán un mayor impacto en los ingresos de la empresa. 

Por ejemplo, es posible que algunos de los artículos no se hayan vendido en las tiendas y supongan una pérdida de dinero. Por lo tanto, tenemos que buscar los artículos que se venden rápidamente en las tiendas.

```python
df_hist_cost = df_hist[["SKU_number", "PriceReg", "ItemCount", "File_Type"]]
```

```python
df_hist_cost["Cost"] = df_hist_cost["PriceReg"] * df_hist_cost["ItemCount"]
```

```python
df_hist_cost.head()
```

```python
df_hist_cost = df_hist_cost.sort_values(by = ["Cost"], ascending = False)
```

```python
df_hist_cost["Accum Cost"] = df_hist_cost["Cost"].cumsum()
```

```python
df_hist_cost["Total Cost"] = df_hist_cost["Cost"].sum()
df_hist_cost["%"] = (df_hist_cost["Accum Cost"]/df_hist_cost["Total Cost"])*100
```

## Análisis ABC

Se toma en cuenta los siguientes porcentajes para determinar las categorías.

1.   A- cuando la columna "%" es menor al 60%
2.   B- cuando la columna "%" esta entre mayor o igual al 60% y menor que el 80%
3.   C- cuando la columna "%" esta entre mayor o igual al 80%


```python
def classes(Porcentajeabc):
    
    if Porcentajeabc>0 and Porcentajeabc<60:
        return "A"
    elif Porcentajeabc>=60 and Porcentajeabc<80:
        return "B"
    else:
        return "C"
```

```python
df_hist_cost["Categoria"] = df_hist_cost["%"].apply(classes)
```
